```python
import numpy as np
import pandas as pd
import numpy as np
import math
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import pydotplus
```

```python
from sklearn import datasets
iris_dataset=datasets.load_iris()
iris_dataset
```

```python
df=pd.DataFrame(iris_dataset.data)
print(df.shape)
```

```python
df.columns=["sl","sw","pl","pw"]
print(df.head())
```

```python
#making label method to categorize different data columns to be ready to fit in our classifier
def Label(val,*par):
    if val<par[0]:
        return "1"
    if val<par[1]:
        return "2"
    if val<par[2]:
        return "3"
    return "4"
#the following method will assign values to new columns made by having value between maximum and minimum for any given feature column
def toLabel(df,feature):
    second=df[feature].mean()
    first=(df[feature].min()+second)/2
    third=(df[feature].max()+second)/2
    return df[feature].apply(Label,args=(first,second,third))
```

```python
#assigining a definite value to new columns based on it's relative value in the dataset
df["sepal_length"]=toLabel(df,"sl")
df["sepal_width"]=toLabel(df,"sw")
df["petal_length"]=toLabel(df,"pl")
df["petal_width"]=toLabel(df,"pw")
df.head()
```

```python
#now as we have categorized our feature columns having defininte data, we can drop existing feature columns 
df.drop(["sl","sw","pl","pw"],axis=1,inplace=True)
set(df["sepal_length"])
df.head()
```

```python
#having target data out of dataset so that its value could be used to fit the data into classifier and move upon score of the model
targ= pd.DataFrame(iris_dataset.target)
unused_features = df.columns
col_names=iris_dataset.target_names
```

```python
#having X_train and Y_train respectively to fit into our Decision Tree Classifier
X_train=df.values
Y_train=targ.values
clf=DecisionTreeClassifier()
clf.fit(X_train,Y_train)
```

```python
#now generating dot_data for our classifier such that we can ultimately have our decision tree be given as result 
dot_data=export_graphviz(clf,out_file=None,filled=True,rounded=True,special_characters=True,feature_names=iris_dataset.feature_names,class_names=iris_dataset.target_names)
graph=pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf("iris_decision_tree.pdf")
```
